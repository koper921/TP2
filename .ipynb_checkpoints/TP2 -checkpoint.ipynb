```python
#Q5 on remarque du bruit , le module la photo de couleur a l'air de contenir des couleurs ...



import cmath
cmath.rect(mod_i1[0][0][0] , arg_i1[0][0][0] )

def apply_cmath_rect( module, argument  ):
    dim1, dim2, dim3 = argument.shape 
    complex_matrix = np.zeros(argument.shape, dtype= complex )
    for i in range(dim1):
        for j in range(dim2):
            for k in range(dim3):
                complex_matrix[i][j][k] = cmath.rect(module[i][j][k] , argument[i][j][k])
    return complex_matrix
essai = apply_cmath_rect(mod_i1, arg_i1)  # donne donc la meme valeur que image1_tf
#print(essai[0])
#print(image1_tf[0])

im_croise1 = apply_cmath_rect(mod_i1, arg_i2)
im_croise2 = apply_cmath_rect(mod_i2, arg_i1)

print("module 1 * arg 2 ")
plt.imshow(np.abs(im_croise1))
plt.show()



plt.imshow(np.abs(np.angle(im_croise1)))
plt.show()

plt.imshow(np.abs(np.real(im_croise1)))
plt.show()






print("module 2 * arg 1")
plt.imshow(np.abs(im_croise2))
plt.show()
plt.imshow(np.abs(np.angle(im_croise2)))
plt.show()
plt.imshow(np.abs(np.real(im_croise2)))
plt.show()


```

```python
#Q6
#filtre moyenne   , moyenne glissante par rapport  N 
# plus on augmente N plus l image se floute , c 'est normale car on fait la moyenne des pixels sur matrice de dimension N , 
# plus la matrice est grande plus on perd de l'information 



def carre(x):
    return x*x

def conv_I_g (image, N):
    one = np.ones((N,N))
    conv = image
    dim1, dim2 =image.shape
    s = 0
    for i in range(N, dim1-N):
        for j in range(N, dim2- N):
            s=0
            for n1 in range(N):
                for n2 in range(N):
                    s = s + image[i -n1, j -n2]
            s = s *1/carre(N)
            conv[i,j ] = s   
            
            #conv_I_g[i,j ] = image[i-N: N+i , j-N: N+j ].dot(one)*1/N**2
    plt.imshow(conv)
    plt.show()
    return conv_I_g
f_prime =misc.imread('house.png',flatten=False, mode= 'F')
conv_I_g(f_prime, 3)
f_prime =misc.imread('house.png',flatten=False, mode= 'F')

plt.imshow(f_prime)
plt.show()
```
