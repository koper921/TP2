```python
Différence en zoom zero padding et DCT:
    zoom zero padding fait apparaitre des oscillations sur le coté de l image (effet de gibbs), 
    cela est dut au fait qu on ne garde que les basses fréquences, et l image subbit des pertes de 
    contraste
    le fait de n avoir garder que les basses fréquences (car haute fréquence = 0)
    ne change pas l amplitude des oscillations
    mais celles ci ne s amortissent pas suffisament vites et les rendent visibles
    les oscillations ont lieu au voisinage des dicontinuités
    (les discontinuités du signal correspondent aux hautes fréquences dans le plan de fourier )

    Zoom par DCT remédit à l effet de Gibbs
    lorsque l on symetrise l image on symetrise le signal, le signal est alors pair, la tranformation
    de fourier ne fait alors qu intervenir que les cosinus (DCT)
    Cette tranformation permet de supprimer l effet de Gibs au bord de l image
    car le fait de symetriser permet de réduire les fortes discontinuités au niveau des contours
    moins de discontinuité => moins d oscilations
    
        
```

```python
from scipy import misc
import numpy as np
f = misc.imread('house.png', flatten=False, mode='RGB')
f_alpha=  misc.imread('house.png', flatten=False, mode='RGBA')
f_prime =misc.imread('house.png',flatten=False, mode= 'F')

f = f/255
f_alpha =f_alpha/255
f_prime =f_prime/255

print(f.shape)
print(f_prime.shape)
import matplotlib.pyplot as plt
plt.imshow(f)
plt.show()

plt.imshow(f_prime)
plt.show()
```

```python

def bord(image, bord):
    if len(image.shape) == 3:
        image =np.mean(image,axis=2)
    image2 = np.zeros((image.shape[0]+2*bord, image.shape[1]+2*bord), dtype=complex)
    image2[bord:-bord,bord:-bord]= image
    return image2 
    
```

```python
from numpy.fft import fft2, fft , ifft2, fftshift, ifftshift
import numpy as np
def zero_padding(image ):
    plt.imshow(image)
    plt.show()
    image =image/255
    image_tf= np.zeros(image.shape, dtype= complex)
    image_tf = np.fft.fft2(image)
    i= fftshift(image_tf)
    im = ifft2(ifftshift(bord(i,3000)))
    
    plt.imshow(np.abs(im))
    plt.show()
    return im 
    
zero_padding(f_prime)    
```

```python
import numpy as np
import matplotlib.image as mpimg
from math import sqrt



def zoom_arr(image):

    #Show image
    plt.imshow(image)
    plt.show()
    dim1, dim2 = image.shape

    zoom_factor=2
    mu=15
    sigma=0.8*sqrt(zoom_factor**2-1)

    I=fftshift(fft2(image))
    I=I*np.random.normal(mu, sigma, (dim1,dim2))

    I_dezoom=ifft2(I).real
    I_ZOOM_OUT=I_dezoom[::zoom_factor,::zoom_factor]

    plt.imshow(I_ZOOM_OUT)
    
    plt.show()
    return I
zoom_arr(f_prime)
```

```python
def miroir (image ):
    image_sim = np.zeros(image.shape)
    dim1, dim2 =image.shape
    centre_sim = dim2/2
    for i in range(dim1):
        for j in range(dim2):
            if j <=int(dim2/2):
                image_sim[i,j]= image[i,j]
            #image_sim[i,dim2-1-j]= image[i,-j+dim2-1]
            elif j >= int(dim2/2)+1:
                image_sim[i, j]= image[i,-j+dim2-1]  #effet miroir 
    plt.imshow(image_sim)
    plt.show()
    return image_sim



from scipy import fftpack

def zoom_dct(image):
    
    i = np.fft.fft2(miroir(image))
    i2= fftshift(i)
    i3 = ifft2(ifftshift(bord(i2,50)))
    dim1,dim2= i3.shape
    i4=np.zeros((int(dim1/2),int(dim2/2) ))
    i4= i3[:int(dim1/2), :int(dim2/2)]
    plt.imshow(np.abs(i4.real))
    plt.show()
    return i4
zoom_dct(f_prime)
```
